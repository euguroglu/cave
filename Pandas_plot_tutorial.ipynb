```
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```
sns.set_style('darkgrid')
```

```
df1 = pd.read_csv('df1',index_col=0)
```

```
df1.head()
```

```
df2 = pd.read_csv('df2')
```

```
df2.head()
```

## Histrogram

Method 1

```
df1['A'].hist()
```

Method 2

```
df1['A'].plot(kind='hist')
```

## Area plot

```
df2.plot.area()
```

```
df2.plot.area(alpha=0.5) #adding transparency
```

## Bar plot

```
df2.plot.bar(figsize=(12,8))
```

```
df2.plot.bar(stacked=True,figsize=(12,8))
```

## Line plot

```
df1.plot.line(y='B',figsize=(12,8))
```

```
df1.index
```

## Scatter plot

```
df1.plot.scatter(x='A',y='B',c='C',figsize=(12,8),cmap='coolwarm')
```

```
df1.plot.scatter(x='A',y='B',s=df1['C']*100,figsize=(12,8))
```

## Box plot

```
df2.plot.box(figsize=(12,8))
```

```
df = pd.DataFrame(np.random.randn(1000,2),columns = ['a','b'])
```

```
df.plot.hexbin(x='a',y='b',figsize=(12,8),gridsize=25)
```

## KDE plot

```
df2['a'].plot.kde(figsize=(12,8))
```
